```python
from cmeutils.dynamics import msd_from_gsd
from cmeutils.structure import get_centers, dihedral_distribution


import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import signac


import os
```

```python
color_dict = {1.0: "#0F0E0F", 0.9: "#450B0C", 0.8: "#7B0708", 0.7: "#B00405", 0.6: "#E60001"}
colors = ["#0F0E0F", "#7B0708", "#E60001", ]

plt.rcParams["figure.figsize"] = (3, 3)
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["legend.fontsize"] = 4
plt.rcParams['lines.markersize'] = 2
plt.rcParams["lines.linestyle"] = "-"
plt.rcParams["lines.marker"] = "o"
plt.rcParams["lines.linewidth"] = 1
plt.rcParams["figure.dpi"] = 1200

gradient = cm.get_cmap('coolwarm', 20)
newcolors = gradient(np.linspace(0, 1, 20))
```

```python
project = signac.get_project("../pekk-tg/polybinder-flow/")
msd_start = -50
msd_stop = -1
```

```python
ua_D = []
ua_temps = []
for kT, job in project.find_jobs({"remove_hydrogens": True}).groupby("kT_quench"):
    ua_temps.append(kT)
    job = list(job)[0]
    msd = msd_from_gsd(
        gsdfile=job.fn("sim_traj.gsd"), start=msd_start, stop=msd_stop, atom_types="all",
    )
    slope = (msd.msd[-5] - msd.msd[-25]) / 20
    ua_D.append(slope)
    plt.plot(msd.msd, label=kT, markersize=1, linewidth=1)

plt.legend()
plt.show()
```

```python
aa_D = []
aa_temps = []

for kT, job in project.find_jobs({"remove_hydrogens": False}).groupby("kT_quench"):
    aa_temps.append(kT)
    job = list(job)[0]
    msd = msd_from_gsd(
        gsdfile=job.fn("sim_traj.gsd"), start=msd_start, stop=msd_stop, atom_types="all",
    )
    slope = (msd.msd[-5] - msd.msd[-25]) / 20
    aa_D.append(slope)
    plt.plot(msd.msd, label=kT, markersize=1, linewidth=1)

plt.legend()
plt.show()
```

```python
fig = plt.figure()
plt.plot(ua_temps, ua_D, label="UA")
plt.plot(aa_temps, aa_D, label="AA")

plt.legend()
plt.show()
```
