```python
import numpy as np
import matplotlib.pyplot as plt
import signac
```

```python
project = signac.get_project("../../coarse-grain-sims/annealing/dt-sweep/")
```

```python
fig = plt.figure(figsize=(20, 10))
for dt, job in project.find_jobs().groupby("dt"):
    job = list(job)[0]
    if job.sp.dt > 0.001:
        continue
    data = np.genfromtxt(job.fn("sim_traj.txt"), names=True)
    kT = data["mdcomputeThermodynamicQuantitieskinetic_temperature"]
    plt.plot(kT[-500:], label=job.sp.dt)
    print(job.sp.dt, np.mean(kT[-1000:]))
    
plt.legend()
```
