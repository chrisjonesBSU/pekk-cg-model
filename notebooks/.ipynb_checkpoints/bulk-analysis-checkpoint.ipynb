# Specific Volume: UA System <a name="specvol"></a>

```python
for kT, jobs in ua_npt_project.find_jobs(
    filter=None, doc_filter={"done": True}).groupby("kT_quench"):
    if kT == 2.8:
        continue
    job = list(jobs)[0]
    data = np.genfromtxt(job.fn("sim_traj.txt"), names=True)
    vol = data["mdcomputeThermodynamicQuantitiesvolume"][-2000:]
    vol_ang = np.mean(vol) * (job.doc.ref_distance**3)
    vol_cm = vol_ang * ((1e-8)**3)
    
    with gsd.hoomd.open(job.fn("init.gsd")) as traj:
        total_mass_red = np.sum(traj[0].particles.mass)
        total_mass_amu = total_mass_red * job.doc.ref_mass
        total_mass_g = total_mass_amu * 1.66e-24
    
    density = total_mass_g / vol_cm
    exp_temp = job.doc.T_SI - 273.15
    job.doc.T_celsius = exp_temp
    job.doc.specific_vol = 1/density
    job.doc.average_density = density
```

```python
fig = plt.figure()

temps = []
density = []

below_tg_x = []
below_tg_y = []
above_tg_x = []
above_tg_y = []

for kT, jobs in ua_npt_project.find_jobs(
    filter=None, doc_filter={"done": True}).groupby("kT_quench"):
    if kT >= 5.8 or kT < 3.2:
        continue
    job = list(jobs)[0]
    plt.plot(job.doc.T_celsius, job.doc.average_density, "k", marker="o", markersize=3)
    temps.append(job.doc.T_celsius)
    density.append(job.doc.average_density)
    if job.doc.T_celsius <= 165:
        below_tg_x.append(job.doc.T_celsius)
        below_tg_y.append(job.doc.average_density)
    else:
        above_tg_x.append(job.doc.T_celsius)
        above_tg_y.append(job.doc.average_density)

plt.vlines(x=165, ymax=1.5, ymin=1.0, color="orange", ls="--", label="$Exp.\,T_g$")

slope, intercept, _, _, _ = linregress(below_tg_x, below_tg_y)
plt.plot(temps, (np.array(temps) * slope) + intercept, "k--")

slope, intercept, _, _, _ = linregress(above_tg_x, above_tg_y)
plt.plot(temps, (np.array(temps) * slope) + intercept, "k--")

plt.xlabel("T [$^{\circ}C$]")
plt.ylabel(r"$\rho\,[\frac{g}{cm^3}]$")
plt.legend()
plt.tight_layout()
plt.savefig("figures/ua-tg-density.png")
plt.show()
```

# Mean Square Displacements <a name="msds"></a>

## UA System

```python
ua_results = {}
chunk_size=200
start_frame = 500
end_frame = 2000 - chunk_size
n_samples = 20

for kT, job in ua_nvt_project.find_jobs().groupby("kT_quench"):
    job = list(job)[0]

    with gsd.hoomd.open(job.fn("components.gsd")) as traj:
        end_frame = len(traj) - chunk_size
        random_starts = np.random.randint(start_frame, end_frame, n_samples)
        
    window_slopes = []
    steps_per_frame = job.doc.steps_per_frame
    ts = job.doc.real_timestep * 1e-15
    ts_frame = steps_per_frame * ts
    
    for i in random_starts:
        msd = msd_from_gsd(
            gsdfile=job.fn("components.gsd"),
            start=int(i),
            stop=int(i) + chunk_size,
            atom_types="all",
            msd_mode="direct"
        )
        
        msd_data = np.copy(msd.msd)
        msd_data *= 3.39**2 * 1e-20
        time = np.arange(0, chunk_size, 1) * ts_frame
        mid_chunk = msd_data[len(msd_data)//4:-len(msd_data)//4]
        mid_chunk_time = time[len(time)//4:-len(time)//4]
        window_slopes.append((mid_chunk[-1] - mid_chunk[0])/(mid_chunk_time[-1] - mid_chunk_time[0]))
        
    ua_results[job.doc.T_SI - 273] = [np.mean(window_slopes), np.std(window_slopes)]
    job.doc.D = np.mean(window_slopes)
    job.doc.D_std = np.std(window_slopes)
    job.doc.T_celsius = job.doc.T_SI - 273
```

```python
fig = plt.figure()
for kT, job in ua_nvt_project.find_jobs().groupby("kT_quench"):
    job = list(job)[0]
    if kT >= 5.8 or kT < 3.2:
        continue
    plt.errorbar(
        x=job.doc.T_celsius, y=job.doc.D, yerr=job.doc.D_std, c="k", marker="o", markersize=3
    )

plt.vlines(x=165, ymax=4e-10, ymin=-0.2e-10, color="orange", ls="--", label="$Exp. T_g$")
plt.hlines(y=0, xmin=64, xmax=320, ls="--", color="b", label="$D_{self}=0$")
plt.legend(loc="upper left")
plt.xlabel("T [$^{\circ}C$]")
plt.ylabel("D $[\dfrac{m^2}{s}]$")
plt.tight_layout()
plt.savefig("figures/pekk-ua-msd-tg.png")
```

# Mean Square Displacements (CG System)

```python
chunk_size=300
start_frame = 500
end_frame = 2000 - chunk_size
n_samples = 20
random_starts = np.random.randint(start_frame, end_frame, n_samples)
cg_results = {}

for kT, jobs in cg_project.find_jobs(
    {"cg_potentials_dir": "msibi-shallow"},
    doc_filter={"done": True}).groupby("kT_quench"):
    
    slopes = {}
    for job in jobs:
        try:
            job.doc.D
            continue
        except:
            pass
        steps_per_frame = job.doc.steps_per_frame
        ts = job.doc.real_timestep * 1e-15
        ts_frame = steps_per_frame * ts
        
        window_slopes = []
        for i in random_starts:
            msd = msd_from_gsd(
                gsdfile=job.fn("sim_traj.gsd"),
                start=int(i),
                stop=int(i) + chunk_size,
                atom_types=["K"],
                msd_mode="direct"
            )
            msd_data = np.copy(msd.msd)
            msd_data *= 3.3997**2 * 1e-20
            time = np.arange(0, chunk_size, 1) * ts_frame
            mid_chunk = msd_data[len(msd_data)//4:-len(msd_data)//4]
            mid_chunk_time = time[len(time)//4:-len(time)//4]
            window_slopes.append((mid_chunk[-1] - mid_chunk[0])/(mid_chunk_time[-1] - mid_chunk_time[0]))               
            slopes[job.sp.ekk_weight] = [np.mean(window_slopes), np.std(window_slopes)]
            job.doc.D = np.mean(window_slopes) / 6
            job.doc.D_std = np.std(window_slopes)
            job.doc.T_celsius = job.doc.T_SI - 273
            
    cg_results[job.doc.T_SI - 273] = slopes
```

```python
left_fits = []
right_fits = []
start_temp = []
start_idx = []
start = 5
for t in inv_temps[start:]:
    start_temp.append(t)
    start_idx.append(start)
    left_x = inv_temps[start:]
    left_y = ln_D[start:]
    right_x = inv_temps[0:start]
    right_y = ln_D[0:start]
    
    left_fit = linregress(x=left_x, y=left_y)
    left_fits.append(left_fit[2])
    right_fit = linregress(x=right_x, y=right_y)
    right_fits.append(right_fit[2])
    
    start += 1
    if start == len(inv_temps) - 5:
        break
```

```python

chunk_size=200
start_frame = 400
end_frame = 1000 - chunk_size
n_samples = 20
random_starts = np.random.randint(start_frame, end_frame, n_samples)

steps_per_frame = 300000
ts = 0.435e-15
ts_frame = steps_per_frame * ts
ti_ratio = 0.8

weights = [0.6, 0.7, 0.8, 1.0]
temps = []
slopes = []


for ti_ratio in weights:
    _temps = []
    _slopes = []
    _slopes_std = []
    
    for kT, jobs in cg_project.find_jobs(
        {"cg_potentials_dir": "rmax_4_angles2", "ekk_weight": ti_ratio},
        doc_filter={"done": True}).groupby("kT_quench"):

        job = list(jobs)[0]    
        window_slopes = []
        for i in random_starts:
            msd = msd_from_gsd(
                gsdfile=job.fn("sim_traj.gsd"),
                start=int(i),
                stop=int(i) + chunk_size,
                atom_types="all",
                msd_mode="direct"
            )
            msd_data = np.copy(msd.msd)
            msd_data *= 3.39**2 * 1e-20
            time = np.arange(0, chunk_size, 1) * ts_frame
            mid_chunk = msd_data[len(msd_data)//4:-len(msd_data)//4]
            mid_chunk_time = time[len(time)//4:-len(time)//4]
            window_slopes.append((mid_chunk[-1] - mid_chunk[0])/(mid_chunk_time[-1] - mid_chunk_time[0]))

        _temps.append(kT)
        _slopes.append(np.mean(window_slopes))
        _slopes_std.append(np.std(window_slopes))

    temps.append(_temps)
    slopes.append(_slopes)
```
