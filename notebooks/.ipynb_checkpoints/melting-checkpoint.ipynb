```python
@jit(nopython=True)
def bond_orientations(positions, bond_groups, images, box):
    bond_orientations = []
    for group in bond_groups:
        pos1 = positions[group][0] + (images[group][0] * box)
        pos2 = positions[group][1] + (images[group][1] * box)
        bond_vec = pos2 - pos1
        unit_vec = bond_vec / np.linalg.norm(bond_vec)
        bond_orientations.append((bond_vec[0], bond_vec[1], bond_vec[2], 0))
    return bond_orientations
```

```python
bonds = bond_orientations(
    snap.particles.position,
    snap.bonds.group,
    snap.particles.image,
    snap.configuration.box[0:3]
)
```

```python
def nematic_order_param(gsd_file, start, stop, director=np.array([0, 0, -1])):
    z_director = []
    order = []
    with gsd.hoomd.open(gsd_file) as traj:
        for snap in traj[start:stop]:
            bonds = bond_orientations(
                snap.particles.position, snap.bonds.group, snap.particles.image, snap.configuration.box[0:3]
            )
            nematic = freud.order.Nematic(director)
            nematic.compute(bonds)
            z_director.append(np.sqrt(nematic.director[2]**2))
            order.append(nematic.order)
    return z_director, order
```

```python
def end_to_end(gsd, start, stop, mean=False):
    sys = System(gsd_file=gsd, atoms_per_monomer=1)
    re_values = []
    for i in range(0, n_frames):
        sys.update_frame(frame=-(i+1))
        re = sys.end_to_end_distances()
        re_values.extend(re)
    if mean:
        re_values = np.mean(re_values)
    return re_values
```

```python
temp_frames = np.arange(0, 2000-133, 133)
box_constraint = {'x': 3.636, 'y': 4.602, 'z': None}
density = 1.38
n_compounds = 72
job_filter = {
    "box_constraints": box_constraint,
    "cg_potentials_dir": "rmax_4_angles2",
    "density": density,
    "n_compounds": [n_compounds]
}

frame_index = 5

fig = plt.figure()

for weight, job in project.find_jobs(job_filter).groupby("ekk_weight"):   
    job = list(job)[0]
    sys = System(gsd_file=job.fn("sim_traj.gsd"), atoms_per_monomer=1)
    start_frame = temp_frames[frame_index] + 20
    end_frame = temp_frames[frame_index] + 50
    re_values = []
    for i in range(start_frame, end_frame):
        sys.update_frame(frame=i)
        re = sys.end_to_end_distances()
        re_values.extend(re)
    bin_centers, bin_heights = get_histogram(re_values)
    plt.plot(bin_centers, savitzky_golay(bin_heights, window_size=3, order=1), c=color_dict[weight], label=weight)
    
plt.legend()
plt.xlim(0, 20000)
plt.show()
```
