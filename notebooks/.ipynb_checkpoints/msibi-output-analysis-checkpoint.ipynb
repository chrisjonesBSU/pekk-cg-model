```python
#kek_file = "KEK_optimization_run_all_para/K-E-K_smoothed.txt"
kek_file = "final_bond_potentials/K-E-K.txt"
data = np.loadtxt(kek_file)
plt.plot(data[:,0], data[:,1])
plt.xlim(1.5, 3.15)
plt.ylim(-1, 30)
```

```python
for file in os.listdir("final_bond_potentials/"):
    print(file)
    fig = plt.figure()
    data = np.loadtxt(os.path.join("final_bond_potentials", file))
    plt.plot(data[:,0], data[:,1])
    plt.title(file)
    plt.show()
```
