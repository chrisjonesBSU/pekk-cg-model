# Specific Volume: UA System <a name="specvol"></a>

```python
### Calculate specific volume
```

```python
for kT, jobs in ua_npt_project.find_jobs(
    filter={"tau_p": 0.1}, doc_filter={"done": True}).groupby("kT_quench"):
    if kT == 2.8:
        continue
    job = list(jobs)[0]
    data = np.genfromtxt(job.fn("sim_traj.txt"), names=True)
    vol = data["mdcomputeThermodynamicQuantitiesvolume"][-2000:]
    vol_ang = np.mean(vol) * (job.doc.ref_distance**3)
    vol_cm = vol_ang * ((1e-8)**3)
    
    with gsd.hoomd.open(job.fn("init.gsd")) as traj:
        total_mass_red = np.sum(traj[0].particles.mass)
        total_mass_amu = total_mass_red * job.doc.ref_mass
        total_mass_g = total_mass_amu * 1.66e-24
    
    density = total_mass_g / vol_cm
    exp_temp = job.doc.T_SI - 273.15
    job.doc.T_celsius = exp_temp
    job.doc.specific_vol = 1/density
    job.doc.average_density = density
```

# Mean Square Displacements <a name="msds"></a>

## UA System

```python
ua_results = {}
chunk_size=200
start_frame = 500
end_frame = 2000 - chunk_size
n_samples = 20

for kT, job in ua_nvt_project.find_jobs().groupby("kT_quench"):
    job = list(job)[0]
    with gsd.hoomd.open(job.fn("components.gsd")) as traj:
        end_frame = len(traj) - chunk_size
        random_starts = np.random.randint(start_frame, end_frame, n_samples)
        
    window_slopes = []
    steps_per_frame = job.doc.steps_per_frame
    ts = job.doc.real_timestep * 1e-15
    ts_frame = steps_per_frame * ts
    
    for i in random_starts:
        msd = msd_from_gsd(
            gsdfile=job.fn("components.gsd"),
            start=int(i),
            stop=int(i) + chunk_size,
            atom_types="all",
            msd_mode="direct"
        )
        
        msd_data = np.copy(msd.msd)
        msd_data *= 3.39**2 * 1e-20
        time = np.arange(0, chunk_size, 1) * ts_frame
        mid_chunk = msd_data[len(msd_data)//4:-len(msd_data)//4]
        mid_chunk_time = time[len(time)//4:-len(time)//4]
        window_slopes.append((mid_chunk[-1] - mid_chunk[0])/(mid_chunk_time[-1] - mid_chunk_time[0]))
        
    ua_results[job.doc.T_SI - 273] = [np.mean(window_slopes), np.std(window_slopes)]
    job.doc.D = np.mean(window_slopes)
    job.doc.D_std = np.std(window_slopes)
    job.doc.T_celsius = job.doc.T_SI - 273
```

```python
fig = plt.figure()
for kT, job in ua_nvt_project.find_jobs().groupby("kT_quench"):
    job = list(job)[0]
    plt.errorbar(
        x=job.doc.T_celsius, y=job.doc.D, yerr=job.doc.D_std, c="k", marker="o"
    )
    #plt.errorbar(
    #    x=job.sp.kT_quench, y=job.doc.D, yerr=job.doc.D_std, c="k", marker="o"
    #)

plt.vlines(x=162.5, ymax=8e-10, ymin=0, color="orange", ls="--", label="Exp. Tg")
plt.vlines(x=358, ymax=8e-10, ymin=0, color="red", ls="--", label="Exp. Tm (80/20)")

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ratios = sorted(list(set(labels)))
handles = [by_label[ratio] for ratio in ratios]
plt.legend(handles, ratios, loc="upper left", title="T/I Ratio")
plt.xlabel("Temperature ($C^{\circ}$)")
plt.ylabel("D $(\dfrac{m^2}{s})$")
plt.title("PEKK Diffusivity (UA Model)", fontsize=7)
```

```python
ekk_weight = 1.0

ln_D = []
inv_temps = []
for kT, jobs in cg_project_tscaled.find_jobs(
    {"cg_potentials_dir": "rmax_4_angles2", "ekk_weight": ekk_weight}).groupby("kT_quench"):
    if kT < 1.8:
        continue
    job = list(jobs)[0]
    ln_D.append(np.log(job.doc.D))
    inv_temps.append(1/kT)
    job.doc.ea_cut = 10
```
